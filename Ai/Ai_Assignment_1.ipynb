```python
import time

print("IDS Algorithm")

def print_node(node):
    print(node[0] + " " + node[1] + " " + node[2] + "\n" + node[3] + " " + node[4] + " " + node[5] + "\n" + node[6] + " " + node[7] + " " + node[8] + "\n-----")

start_state = input("Enter start state: ")
goal_state = input("Enter goal state: ")
print("-----")

depth = 0
path_stack = [""]
ids_stack = [start_state]
ids_level = [0]
path_nodes = []
visited_nodes = []
start_time = time.time()

while ids_stack[0] != goal_state:
    parent_node = path_stack.pop(0)
    visited_node = ids_stack.pop(0)
    node_level = ids_level.pop(0)
    print_node(visited_node)

    if node_level == depth and node_level not in ids_level:
        if visited_node != goal_state:
            depth += 1
            path_stack = [""]
            ids_stack = [start_state]
            ids_level = [0]
            path_nodes = []
            visited_nodes = []
            print("Goal state not found at depth " + str(depth - 1) + "\nDepth increased to " + str(depth) + "\n-----")
            continue
        else:
            break

    elif node_level == depth and node_level in ids_level:
        if visited_node != goal_state:
            continue
        else:
            break

    elif node_level < depth:
        blank_index = visited_node.index('0')

        if blank_index == 0:
            temp_node = ""
            temp_node = temp_node + visited_node[1] + '0' + visited_node[2:]
            if temp_node not in visited_nodes:
                path_stack.insert(0, visited_node)
                ids_stack.insert(0, temp_node)
                ids_level.insert(0, node_level + 1)

            temp_node = ""
            temp_node = temp_node + visited_node[3] + visited_node[1:3] + '0' + visited_node[5:]
            if temp_node not in visited_nodes:
                path_stack.insert(1, visited_node)
                ids_stack.insert(1, temp_node)
                ids_level.insert(1, node_level + 1)

        if blank_index == 1:
            temp_node = ""
            temp_node = temp_node + '0' + visited_node[0] + visited_node[2:]
            if temp_node not in visited_nodes:
                path_stack.insert(0, visited_node)
                ids_stack.insert(0, temp_node)
                ids_level.insert(0, node_level + 1)

            temp_node = ""
            temp_node = temp_node + visited_node[0] + visited_node[4] + visited_node[2:4] + '0' + visited_node[5:]
            if temp_node not in visited_nodes:
                path_stack.insert(1, visited_node)
                ids_stack.insert(1, temp_node)
                ids_level.insert(1, node_level + 1)

            temp_node = ""
            temp_node = temp_node + visited_node[0] + visited_node[2] + '0' + visited_node[3:]
            if temp_node not in visited_nodes:
                path_stack.insert(2, visited_node)
                ids_stack.insert(2, temp_node)
                ids_level.insert(2, node_level + 1)

        elif blank_index == 2:
            temp_node = ""
            temp_node = temp_node + visited_node[0] + '0' + visited_node[1] + visited_node[3:]
            if temp_node not in visited_nodes:
                path_stack.insert(0, visited_node)
                ids_stack.insert(0, temp_node)
                ids_level.insert(0, node_level + 1)

            temp_node = ""
            temp_node = temp_node + visited_node[0:2] + visited_node[5] + visited_node[3:5] + '0' + visited_node[6:]
            if temp_node not in visited_nodes:
                path_stack.insert(1, visited_node)
                ids_stack.insert(1, temp_node)
                ids_level.insert(1, node_level + 1)

        elif blank_index == 3:
            temp_node = ""
            temp_node = temp_node + '0' + visited_node[1:3] + visited_node[0] + visited_node[4:]
            if temp_node not in visited_nodes:
                path_stack.insert(0, visited_node)
                ids_stack.insert(0, temp_node)
                ids_level.insert(0, node_level + 1)

            temp_node = ""
            temp_node = temp_node + visited_node[0:3] + visited_node[6] + visited_node[4:6] + '0' + visited_node[7:]
            if temp_node not in visited_nodes:
                path_stack.insert(1, visited_node)
                ids_stack.insert(1, temp_node)
                ids_level.insert(1, node_level + 1)

            temp_node = ""
            temp_node = temp_node + visited_node[0:3] + visited_node[4] + '0' + visited_node[5:]
            if temp_node not in visited_nodes:
                path_stack.insert(2, visited_node)
                ids_stack.insert(2, temp_node)
                ids_level.insert(2, node_level + 1)

        elif blank_index == 4:
            temp_node = ""
            temp_node = temp_node + visited_node[0] + '0' + visited_node[2:4] + visited_node[1] + visited_node[5:]
            if temp_node not in visited_nodes:
                path_stack.insert(0, visited_node)
                ids_stack.insert(0, temp_node)
                ids_level.insert(0, node_level + 1)

            temp_node = ""
            temp_node = temp_node + visited_node[0:4] + visited_node[7] + visited_node[5:7] + '0' + visited_node[8]
            if temp_node not in visited_nodes:
                path_stack.insert(1, visited_node)
                ids_stack.insert(1, temp_node)
                ids_level.insert(1, node_level + 1)

            temp_node = ""
            temp_node = temp_node + visited_node[0:3] + '0' + visited_node[3] + visited_node[5:]
            if temp_node not in visited_nodes:
                path_stack.insert(2, visited_node)
                ids_stack.insert(2, temp_node)
                ids_level.insert(2, node_level + 1)

            temp_node = ""
            temp_node = temp_node + visited_node[0:4] + visited_node[5] + '0' + visited_node[6:]
            if temp_node not in visited_nodes:
                path_stack.insert(3, visited_node)
                ids_stack.insert(3, temp_node)
                ids_level.insert(3, node_level + 1)

        elif blank_index == 5:
            temp_node = ""
            temp_node = temp_node + visited_node[0:5] + visited_node[8] + visited_node[6:8] + '0'
            if temp_node not in visited_nodes:
                path_stack.insert(0, visited_node)
                ids_stack.insert(0, temp_node)
                ids_level.insert(0, node_level + 1)

            temp_node = ""
            temp_node = temp_node + visited_node[0:4] + '0' + visited_node[4] + visited_node[6:]
            if temp_node not in visited_nodes:
                path_stack.insert(1, visited_node)
                ids_stack.insert(1, temp_node)
                ids_level.insert(1, node_level + 1)

            temp_node = ""
            temp_node = temp_node + visited_node[0:2] + '0' + visited_node[3:5] + visited_node[2] + visited_node[6:]
            if temp_node not in visited_nodes:
                path_stack.insert(2, visited_node)
                ids_stack.insert(2, temp_node)
                ids_level.insert(2, node_level + 1)

        elif blank_index == 6:
            temp_node = ""
            temp_node = temp_node + visited_node[0:6] + visited_node[7] + '0' + visited_node[8]
            if temp_node not in visited_nodes:
                path_stack.insert(0, visited_node)
                ids_stack.insert(0, temp_node)
                ids_level.insert(0, node_level + 1)

            temp_node = ""
            temp_node = temp_node + visited_node[0:3] + '0' + visited_node[4:6] + visited_node[3] + visited_node[7:]
            if temp_node not in visited_nodes:
                path_stack.insert(1, visited_node)
                ids_stack.insert(1, temp_node)
                ids_level.insert(1, node_level + 1)

        elif blank_index == 7:
            temp_node = ""
            temp_node = temp_node + visited_node[0:6] + '0' + visited_node[6] + visited_node[8]
            if temp_node not in visited_nodes:
                path_stack.insert(0, visited_node)
                ids_stack.insert(0, temp_node)
                ids_level.insert(0, node_level + 1)

            temp_node = ""
            temp_node = temp_node + visited_node[0:4] + '0' + visited_node[5:7] + visited_node[4] + visited_node[8]
            if temp_node not in visited_nodes:
                path_stack.insert(1, visited_node)
                ids_stack.insert(1, temp_node)
                ids_level.insert(1, node_level + 1)

            temp_node = ""
            temp_node = temp_node + visited_node[0:7] + visited_node[8] + '0'
            if temp_node not in visited_nodes:
                path_stack.insert(2, visited_node)
                ids_stack.insert(2, temp_node)
                ids_level.insert(2, node_level + 1)

        elif blank_index == 8:
            temp_node = ""
            temp_node = temp_node + visited_node[0:7] + '0' + visited_node[7]
            if temp_node not in visited_nodes:
                path_stack.insert(0, visited_node)
                ids_stack.insert(0, temp_node)
                ids_level.insert(0, node_level + 1)

            temp_node = ""
            temp_node = temp_node + visited_node[0:5] + '0' + visited_node[6:8] + visited_node[5]
            if temp_node not in visited_nodes:
                path_stack.insert(1, visited_node)
                ids_stack.insert(1, temp_node)
                ids_level.insert(1, node_level + 1)

        path_nodes.append(parent_node)
        visited_nodes.append(visited_node)

path_nodes.append(path_stack.pop(0))
visited_nodes.append(ids_stack.pop(0))
ids_level.pop(0)
print_node(visited_nodes[-1])
end_time = time.time()

path_cost = 0
index = -1
while path_nodes[index] != "":
  index = visited_nodes.index(path_nodes[index])
  path_cost += 1

print("Time taken: " + str(end_time - start_time) + " seconds")
print("Path cost: " + str(path_cost))
print("Number of nodes visited: " + str(len(visited_nodes)))
print("Remaining stack: ", ids_stack)
```
